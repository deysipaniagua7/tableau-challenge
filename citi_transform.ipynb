### Extract the 2024 Citi Bike Trip Data Data

```python
# Read the July 2023 data into a Pandas DataFrame.
july_2024 = pd.read_csv('Resources/JC-202407-citibike-tripdata.csv')
july_2024.head()
```

```python
# Get a brief summary of the July 2024 DataFrame.
july_2024.info()
```

```python
# Remove rows with null values and create a cleaned dataframe.
cleaned_july_2024 = july_2024.dropna()

# Get a brief summary of the July 2024 DataFrame.
cleaned_july_2024.info()
```

```python
# Read the December 2024 data into a Pandas DataFrame.
december_2024 = pd.read_csv('Resources/JC-202412-citibike-tripdata.csv')
december_2024.head()
```

```python
# Get a brief summary of the July 2024 DataFrame.
december_2024.info()
```

```python
# Remove rows with null values and create a cleaned dataframe.
cleaned_december_2024 = december_2024.dropna()

# Get a brief summary of the July 2024 DataFrame.
cleaned_december_2024.info()
```

```python
# Merge the cleaned data on thhe ride_id.
merged_df = pd.concat([cleaned_july_2023, cleaned_december_2023, cleaned_july_2024, cleaned_december_2024], ignore_index=True)
merged_df
```

```python
# Save merged_df to csv.
merged_df.to_csv('merged_data.csv', index=False)  # index=False prevents writing row indices to the file
```
